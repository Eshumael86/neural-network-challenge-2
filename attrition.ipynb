## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
print(attrition_df.columns)
```

```python
# Create a list of at least 10 column names to use as X data
x_columns = ['Age', 'YearsSinceLastPromotion', 'OverTime', 'DistanceFromHome', 'Education', 'YearsAtCompany', 'WorkLifeBalance', 'StockOptionLevel', 'NumCompaniesWorked', 'JobSatisfaction',]
# Create X_df using your selected columns
x_df = attrition_df[x_columns]
x_df.head()

# Show the data types for X_df
x_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
m_train, m_test, y_train, y_test = train_test_split(x_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
x_df = pd.get_dummies(x_df)
x_df.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(x_df)
# Transform the training and testing data
x_df = scaler.transform(x_df)
# Scale the training and testing data
x_df = scaler.transform(x_df)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
Department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
Department_encoder.fit(y_df[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
Department_train = Department_encoder.transform(y_train[['Department']])
Department_test = Department_encoder.transform(y_test[['Department']])

Department_train
Department_test
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
attrition_encoder.fit(y_df[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
attrition_train = attrition_encoder.transform(y_train[['Attrition']])
attrition_test = attrition_encoder.transform(y_test[['Attrition']])

attrition_train
attrition_test
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data.
x_df.shape[1]
# Create the input layer
input_layer = layers.Input(shape=(x_df.shape[1],))

# Create at least two shared layers
shared_layer1 = layers.Dense(units=64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(units=32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_layer1 = layers.Dense(units=32, activation='relu')(shared_layer2)
department_output = layers.Dense(units=2, activation='softmax')(department_layer1)
# Create the hidden layer
hidden_layer = layers.Dense(units=64, activation='relu')(shared_layer2)

# Create the output layer
output_layer = layers.Dense(units=2, activation='softmax')(hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_layer1 = layers.Dense(units=32, activation='relu')(shared_layer2)
attrition_output = layers.Dense(units=2, activation='softmax')(attrition_layer1)
# Create the hidden layer
hidden_layer = layers.Dense(units=64, activation='relu')(shared_layer2)

# Create the output layer
output_layer = layers.Dense(units=2, activation='softmax')(hidden_layer)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summarize the model
model.summary()
```

```python
# Train the model
train_test_split(x_df, y_df, random_state=1)
```

```python
from math import e
# Evaluate the model with the testing data
evaluations = model.evaluate(x_df, [Department_test, attrition_test])
```
